SHARPE RATIO ANALYSIS OF IT COMPANIES WITH BENCHMARK: NIFTY 50 FOR THE PERIOD 11th NOV 2023-11th NOV 2024

The Sharpe Ratio is a metric used in finance to assess the performance of an investment compared to a risk-free asset, after adjusting for its risk. It was developed by Nobel laureate William F. Sharpe and is widely used to evaluate how well the return of an asset compensates an investor for the risk taken. In essence, the Sharpe Ratio helps determine whether an investment's returns are due to smart decisions or excessive risk

```python
# calculate daily stock_data returns
stock_returns = stock_data.pct_change()

# plot the daily returns
stock_returns.plot(linewidth=2)

# summarize the daily returns
stock_returns.describe()
```

```python
# calculate daily benchmark_data returns
nifty_returns = benchmark_data['NIFTY 50'].pct_change()

# plot the daily returns
nifty_returns.plot()

# summarize the daily returns
nifty_returns.describe()
```

```python
# Remove duplicate indices
stock_returns = stock_returns[~stock_returns.index.duplicated(keep='first')]
nifty_returns = nifty_returns[~nifty_returns.index.duplicated(keep='first')]

# Calculate excess returns
excess_returns = stock_returns.sub(nifty_returns, axis=0)

# Align both DataFrames to a common date range
common_index = stock_returns.index.intersection(nifty_returns.index)
stock_returns = stock_returns.reindex(common_index)
nifty_returns = nifty_returns.reindex(common_index)

# Calculate excess returns
excess_returns = stock_returns.sub(nifty_returns, axis=0)
```

```python
# calculate the difference in daily returns
excess_returns = stock_returns.sub(nifty_returns,axis=0)

# plot the excess_returns
excess_returns.plot()

# summarize the excess_returns
excess_returns.describe()
```

```python
# calculate the mean of excess_returns 
avg_excess_return = excess_returns.mean()

# plot avg_excess_returns
avg_excess_return.plot.bar(title='Mean of the Return Difference')
```

```python
# calculate the standard deviations
sd_excess_return = excess_returns.std()

# plot the standard deviations
sd_excess_return.plot.bar(title='Standard Deviation of the Return Difference')
```

```python
# calculate the daily sharpe ratio
daily_sharpe_ratio = avg_excess_return.div(sd_excess_return)

# annualize the sharpe ratio
annual_factor = np.sqrt(252)
annual_sharpe_ratio = daily_sharpe_ratio.mul(annual_factor)

# plot the annualized sharpe ratio
annual_sharpe_ratio.plot.bar(title='Annualized Sharpe Ratio: Stocks vs Nifty50')
```

The difference in Sharpe Ratios between these stocks is driven more by their return performance rather than significant differences in risk, with HCL Tech outperforming mainly due to higher average daily returns.

The Sharpe Ratio allows investors to evaluate these stocks on an equal basis, comparing their performance relative to risk in an 'apples-to-apples' way. When faced with such investment options, the Sharpe Ratio provides insight into which stock may yield better returns for the risk assumed.
